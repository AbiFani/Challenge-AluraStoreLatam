## Preparar datos para mapeo
Extraer las columnas de latitud y longitud de los dataframes de cada tienda y combinarlas.


```python
#importar la biblioteca de folium.
import folium
from folium.plugins import HeatMap

# Extraer latitud y longitud y eliminar nulos para cada tienda
ubicaciones_tienda1 = tienda1[['lat', 'lon']].dropna()
ubicaciones_tienda2 = tienda2[['lat', 'lon']].dropna()
ubicaciones_tienda3 = tienda3[['lat', 'lon']].dropna()
ubicaciones_tienda4 = tienda4[['lat', 'lon']].dropna()

# Concatenar todas las ubicaciones en un solo DataFrame
todas_las_ubicaciones = pd.concat([
    ubicaciones_tienda1,
    ubicaciones_tienda2,
    ubicaciones_tienda3,
    ubicaciones_tienda4
])

# Imprimir las primeras 5 filas de cada DataFrame de ubicaciones
print("Ubicaciones Tienda 1:")
print(ubicaciones_tienda1.head())
print("\nUbicaciones Tienda 2:")
print(ubicaciones_tienda2.head())
print("\nUbicaciones Tienda 3:")
print(ubicaciones_tienda3.head())
print("\nUbicaciones Tienda 4:")
print(ubicaciones_tienda4.head())
```

Ahora crearé un mapa base para cada tienda, centrado en la latitud y longitud promedio de sus puntos de venta. Esto implicará calcular las coordenadas medias de cada tienda e inicializar un mapa Folium para cada una.

## Crear mapas base para cada tienda

Inicializar un mapa Folium para cada una de las cuatro tiendas, centrándolos en una ubicación geográfica relevante, como el promedio de latitud y longitud de todas las ventas de la tienda.


```python
centro_tienda1 = [ubicaciones_tienda1['lat'].mean(), ubicaciones_tienda1['lon'].mean()]
centro_tienda2 = [ubicaciones_tienda2['lat'].mean(), ubicaciones_tienda2['lon'].mean()]
centro_tienda3 = [ubicaciones_tienda3['lat'].mean(), ubicaciones_tienda3['lon'].mean()]
centro_tienda4 = [ubicaciones_tienda4['lat'].mean(), ubicaciones_tienda4['lon'].mean()]

print("Centro Tienda 1:", centro_tienda1)
print("Centro Tienda 2:", centro_tienda2)
print("Centro Tienda 3:", centro_tienda3)
print("Centro Tienda 4:", centro_tienda4)
```

Se han calculado las coordenadas promedio de cada tienda. Ahora, inicializaré un mapa de Folium para cada tienda, centrándolo en estas coordenadas promedio calculadas y estableciendo un nivel de zoom adecuado.

```python
mapa_tienda1 = folium.Map(location=centro_tienda1, zoom_start=2)
mapa_tienda2 = folium.Map(location=centro_tienda2, zoom_start=2)
mapa_tienda3 = folium.Map(location=centro_tienda3, zoom_start=2)
mapa_tienda4 = folium.Map(location=centro_tienda4, zoom_start=2)

print("Mapas inicializados para cada tienda.")
```

Se han inicializado los mapas de cada tienda. Para verificarlo, mostraré cada mapa.
